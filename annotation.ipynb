## Annotations loaded from Google drive 

## Step 0: Environment set up

```python
from conn import connect2Googlesheet2, get_relevant_documents
import pandas as pd  # type: ignore

pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

## Step 1: Load data into a dataframe

```python
spreadsheet = connect2Googlesheet2()

# Select the worksheet: relevance
worksheet = spreadsheet.get_worksheet(4)

# Get all records as a list of dictionaries
data = worksheet.get_all_records()

# Convert to Pandas DataFrame
df_relevance = pd.DataFrame(data)
```

## Step 2: Keep relevant docs for each question

```python
relevant_docs_all = {}
for question in df_relevance.columns[1:]:  # Skip the first column which is 'Document'
    relevant_docs_all[question] = get_relevant_documents(df_relevance, question)
# Convert the sets to lists and then join them into strings
for question in relevant_docs_all:
    relevant_docs_all[question] = ", ".join(relevant_docs_all[question])

# Convert the dictionary to a DataFrame
relevant_docs_df = pd.DataFrame(
    list(relevant_docs_all.items())[2:38], columns=["Ques", "Docs"]
)

# save relevant_docs_df as dataframe
relevant_docs_df.to_csv("./outputs/relevant_docs.csv", index=True)
relevant_docs_df
```
